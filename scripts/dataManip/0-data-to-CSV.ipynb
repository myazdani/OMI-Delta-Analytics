```python
import string
import re
import pandas as pd
```

# Cleaning up header description file

```python
column_labels_filepath = "../../data/Basecamp Download/Teamsters_data_layout_plain.txt"

f = open(column_labels_filepath)
lines = [filter(lambda x: x in string.printable, line).strip() for line in f]
f.close()
```

```python
lines
```

Find splits in header description file 

```python
break_indices = []
for i, line in enumerate(lines):
    if len(line) == 0:
        break_indices.append(i)
print break_indices
```

```python
claims_header = [] 
for line in lines[1:break_indices[0]]:
    claims_header.append(line)
print "number of columns for claims:", len(claims_header)
```

```python
patientInfo_header = [] 
for line in lines[break_indices[0]+2:break_indices[1]]:
    patientInfo_header.append(line)
print "number of columns for patient info:", len(patientInfo_header)
```

```python
RXclaims_header = [] 
for line in lines[break_indices[1]+2:]:
    RXclaims_header.append(line)
print "number of columns for RX claims:", len(RXclaims_header)
```

# Create "clean" headers: remove white space and other funky characters

```python
claims_header = [re.sub(r'\W+', '', s.replace("#", "NUM")) for s in claims_header]
patientInfo_header = [re.sub(r'\W+', '', s.replace("#", "NUM")) for s in patientInfo_header]
RXclaims_header = [re.sub(r'\W+', '', s.replace("#", "NUM")) for s in RXclaims_header]
```

# Create clean claims data

```python
claims_path = "../../data/Basecamp Download/DHS_20140801-20150731_JUL_Claims_1.txt"

claimsDF = pd.read_csv(claims_path, sep='\t', header=None)
```

```python
claimsDF.head()
```

```python
claimsDF.columns = claims_header
```

```python
claimsDF.head()
```

# Create clean Patient Info data

```python
patient_path = "../../data/Basecamp Download/DHS_20140801-20150731_JUL_Elig_1.xlsx"

patientDF = pd.read_excel(patient_path, header = None)
```

#Create clean RX claims data

```python
RXclaims_path = "../../data/Basecamp Download/DHS_20140801-20150731_JUL_RX_1.txt"

RXclaimsDF = pd.read_csv(RXclaims_path, sep='\t', header=None)
```
